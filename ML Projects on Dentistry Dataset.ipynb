# Confusion matrix(Accuracy), ROC curve and AUC curve

```python
from sklearn import metrics
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report,auc,accuracy_score,roc_curve,RocCurveDisplay,roc_auc_score
```

```python
# For Logistic Regression

prediction = model_log.predict(X_test) 
conf_m = ConfusionMatrixDisplay(confusion_matrix(y_test,prediction))
conf_m.plot()
print(classification_report(y_test, prediction))
accuracy = accuracy_score(y_test,prediction)
print('Accuracy:', accuracy)
# ROC Curve and AUC
y_prob = model_log.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, prediction)
roc_auc = auc(fpr, tpr)
```

```python
RocCurveDisplay.from_estimator(model_log, X_test, y_test)
```

```python
RocCurveDisplay.from_estimator(model_dec, X_test, y_test)
```
